The implementation was simple but not very modular since the loss and gradient were computed in a single monolithic function. This is manageable for a simple two-layer network, but would become impractical as we move to bigger models. Ideally we want to build networks using a more modular design so that we can implement different layer types in isolation and then snap them together into models with different architectures.

## Inline Question 1:

We've only asked you to implement ReLU, but there are a number of different activation functions that one could use in neural networks, each with its pros and cons. In particular, an issue commonly seen with activation functions is getting zero (or close to zero) gradient flow during backpropagation. Which of the following activation functions have this problem? If you consider these functions in the one dimensional case, what types of input would lead to this behaviour?
1. Sigmoid
2. ReLU
3. Leaky ReLU

## Answer:

1. In the sigmoid function, a significant issue arises when a neuron's activation approaches either extreme end of the curve, reaching values very close to 0 or 1. At these points, the gradient of the function becomes almost zero. This occurs when the input is an extremely negative number, like -100,000 (resulting in a sigmoid value very close to 0), or an extremely positive number, such as 100,000 (resulting in a sigmoid value very close to 1). When the gradient is almost zero, it effectively "kills" the gradient flow. Consequently, very little signal propagates through the neuron to update its weights and, subsequently, the entire network.

2. In the case of the ReLU activation function, the gradient behaves as 1 for positive input values (x > 0) and 0 for negative input values (x < 0). When all inputs to a ReLU neuron are negative, it causes the vanishing gradient problem. This issue is commonly referred to as the "dying ReLU" problem, where certain neurons effectively become inactive and never update their weights. In a one-dimensional scenario, this can be illustrated by an input vector consisting entirely of negative values, such as [-10, -10, -10, -10].

3. Leaky ReLU is designed to address the "dying ReLU" problem. It introduces a small slope to the negative part of the curve, preventing complete inactivity of neurons for negative inputs. However, it's worth noting that a poor initialization, such as an all-zero vector, can still lead to issues in training.
